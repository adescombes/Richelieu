## Rue richelieu

### Extraction des personnes

```python
street_name = 'richelieu'
```

```python
df_geo[df_geo['NOM_ENTIER,C,50'].str.lower().str.contains(street_name)]['NOM_ENTIER,C,50'].unique()
```

```python
negatives = ['neuve', 'nve']
```

```python
np.logical_and.reduce([df['Rue'].str.lower().str.contains('%s'%street_name)] + [~df['Rue'].str.lower().str.contains('%s'%neg) for neg in negatives]).sum()
```

```python
distances_target = []
distances_negatives = []
for items in tqdm(df['Rue'].str.lower().str.split()):
    dists_target = []
    dists_negatives = []
    for item in items:
        dists_target.append(editdistance.eval(street_name, item))
        for neg in negatives:
            dists_negatives.append(editdistance.eval(neg, item))
    distances_target.append(min(dists_target))
    distances_negatives.append(min(dists_negatives))
distances_target = np.array(distances_target)
distances_negatives = np.array(distances_negatives)
```

```python
print('Il y a %d personnes ayant une rue contenant le mot richelieu a une distance de 2'%(distances_target < 3).sum())
print('Il y a %d personnes ayant une rue contenant le mot richelieu a une distance de 2 et ne contenant pas le mot "nve" ou "neuve"'%((distances_target < 3) & (distances_negatives > 1)).sum())
print('Il y a %d personnes ayant une rue contenant le mot richelieu a une distance de 2 et ne contenant pas le mot "nve" ou "neuve" en ayant moins de 3 mots dans le champ'%(((distances_target < 3) & (distances_negatives > 1) & (df['Rue'].str.split().apply(len) <= 3)).sum()))
```

```python
df_curr_street = df[(distances_target < 3) & (distances_negatives > 1) & (df['Rue'].str.split().apply(len) <= 3)]
curr_count = len(df_curr_street)
print("Nous avons %d personnes"%curr_count)
```

### Réalignement sur Vasserot

```python
df_curr_street_realign_num = df_curr_street.copy()
df_curr_street_realign_num['num_clean'] = df_curr_street_realign_num['Numéro'].str.replace('[^0-9\s]', ' ').str.strip()
df_curr_street_realign_num = df_curr_street_realign_num[df_curr_street_realign_num['num_clean'].str.split().apply(len) == 1]
print("Nous avons maintenant %d personnes, nous avons éliminé %d personnes (%.2f%%)"%(len(df_curr_street_realign_num), curr_count-len(df_curr_street_realign_num), (1-len(df_curr_street_realign_num)/curr_count)*100))
```

```python
num_realign = df_curr_street_realign_num['num_clean'].isin(df_geo[df_geo['NOM_ENTIER,C,50'].str.contains('Rue de Richelieu')]['NUM_VOIES,C,10']).sum()
print("En réalignant les personnes sur le plan Vasserot, nous avons %d personnes, nous avons éliminé %d personnes (%.2f%%)"%(num_realign, curr_count-num_realign, (1-num_realign/curr_count)*100))
```
